```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies.shape
```

```python
movies = movies[['movie_id','title','genres','keywords','overview','cast','crew']]
```

```python
movies.head(1)
```

```python
movies['genres'][0]
```

```python
import ast
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['crew'][0]
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
import ast
def convert_cast(obj):
  l=[]
  counter = 0;
  for i in ast.literal_eval(obj):
    if counter !=3:
      l.append(i['name'])
    else:
      break
  return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
import ast
def convert_crew(obj):
  l=[]
  counter = 0;
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies.dropna(inplace=True)
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head(1)
```

```python
movies['cast']=movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies['tag'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df = movies[['movie_id','title','tag']]
new_df
```

```python
new_df['tag']=new_df['tag'].apply(lambda x : " ".join(x))
```

```python
new_df
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  l=[]
  for i in text.split():
    l.append(ps.stem(i))
  return " ".join(l)
```

```python
new_df['tag']=new_df['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(new_df.iloc[:,-1]).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
r = sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x :x[1])[1:6]
r
```

```python
r[0][0]
```

```python
def reccomnd(movie):
  movie_index = new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  reccomended = sorted(list(enumerate(distances)),reverse=True,key=lambda x :x[1])[1:6]

  for i in reccomended:
    print(new_df.iloc[i[0],1])
```

```python
reccomnd('Avatar')
```
